```python
path = '/home/heyang/data/IV/IVResult/'
Table = []
colname = []
for var in ConVar:
    data = pd.read_csv(path+var+'.csv')
    cutpoint = list(data.Cutpoint)
    cutpoint = cutpoint[0:len(cutpoint)-2]
    cutpoint = [float(x.split(' ')[1]) for x in cutpoint]
    woe = list(data.WoE)
    woe = woe[0:len(woe)-2]
    varname = dic_num_val[var]
    colname.append(varname+'_cutpoint')
    colname.append(varname+'_woe')
    Table.append(cutpoint)
    Table.append(woe)
```
